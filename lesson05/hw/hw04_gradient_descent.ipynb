## 梯度下降法 (Gradient Descent)

```python
# 載入套件
import numpy as np
import matplotlib.pyplot as plt
```

```python
def GD(w_start, df, epochs, lr):
    """
    梯度下降函數  
    `w_start`: w 的起始點  
    `df`: 目標函數的一階導數  
    `epochs`: 執行週期  
    `lr`: 學習率
    """
    w_list = np.zeros(epochs+1)
    w = w_start
    w_list[0] = w

    for i in range(epochs):
        # w 更新 w_new = w - learning_rate * gradient        
        w -= lr * df(w) 
        w_list[i+1] = w

    return w_list
```

```python
def plot_GD(f, w):
    """
    繪製曲線
    """   
    t = np.arange(-6.0, 6.0, 0.01)
    plt.plot(t, [f(x) for x in t], c='b')
    plt.plot(w, [f(x) for x in w], c='r')    
    plt.scatter(w, [f(x) for x in w], c='r') 

    plt.title('Gradient Descent')
    plt.xlabel('W')
    plt.ylabel('Loss')
    plt.show()
```

## 求最小值

### $x^3-2x+100$

```python
# 目標函數 (損失函數)
f = lambda x: x ** 3 - 2 * x + 100
# 目標函數的一階導數
df = lambda x: 3 * x ** 2 - 2

# 超參數 (Hyperparameters)
w_start = 5   # 起始權重
epochs = 300  # 執行週期數 
lr = 0.003    # 學習率

# 梯度下降法 
w = GD(w_start, df, epochs, lr)
print('w的變化軌跡：', np.around(w, 2))
plot_GD(f, w)
```

### $2x^4-3x-20$

```python
# 目標函數 (損失函數)
f = lambda x: 2 * x ** 4 - 3 * x - 20 
# 目標函數的一階導數
df = lambda x: 8 * x ** 3 - 3

# 超參數 (Hyperparameters)
w_start = 5   # 起始權重
epochs = 300  # 執行週期數 
lr = 0.002    # 學習率

# 梯度下降法 
w = GD(w_start, df, epochs, lr)
print('w的變化軌跡：', np.around(w, 2))
plot_GD(f, w)
```

### $\sin(x)e^{-0.1(x-0.6)^2}$

```python
from sympy import *
import math

x = Symbol('x')
y = sin(x) * E ** (-0.1 * (x - 0.6) ** 2)
yprime = y.diff(x)

# 目標函數
def f(x_value): return math.sin(x_value) * math.e ** (-0.1 * (x_value - 0.6) ** 2)
# 目標函數一階導數
df = lambda x_value: yprime.subs(x, x_value).evalf()

# 超參數 (Hyperparameters)
w_start = -3  # 起始權重
epochs = 300  # 執行週期數
lr = 0.03     # 學習率

# 梯度下降法
w = GD(w_start, df, epochs, lr)
print('w的變化軌跡：', np.around(w, 2))
plot_GD(f, w)
```
