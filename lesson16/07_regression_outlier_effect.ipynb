# 迴歸缺點

## 載入測試資料

```python
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=10, n_features=1, noise=20, random_state=123)
```

## 繪圖

```python
from matplotlib import pyplot as plt


fig = plt.figure(figsize=(4, 4))
plt.scatter(X, y, color="blue", alpha=0.5)
plt.title("test data")
```

## 模型訓練

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
lr.coef_, lr.intercept_
```

```python
fig = plt.figure(figsize=(4, 4))
plt.scatter(X, y, color="blue", alpha=0.5)
plt.plot([-2, 2], [-2 * lr.coef_ + lr.intercept_, 2 * lr.coef_ + lr.intercept_], c="r")
plt.title("test data")
```

## 製造離群值

```python
y[0]
```

```python
# 製造離群值
y[0] += 2000
```

## 模型訓練

```python
from sklearn.linear_model import LinearRegression

lr2 = LinearRegression()
lr2.fit(X, y)
lr2.coef_, lr2.intercept_
```

## 繪圖比較

```python
from matplotlib import pyplot as plt
import numpy as np


fig = plt.figure(figsize=(8, 8))
plt.scatter(X, y, color="blue", alpha=0.5)

line_X = np.array([-3, 3])
plt.plot(line_X, line_X * lr.coef_ + lr.intercept_, c="green", label="old regression line")
plt.plot(line_X, line_X * lr2.coef_ + lr2.intercept_, c="red", label="new regression line")
plt.title("test data")
plt.legend()
```
