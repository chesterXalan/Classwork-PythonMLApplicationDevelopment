# 依據屬性資料預測年所得是否超過US$50K/Year

## 載入相關套件

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. 載入資料集

```python
df = pd.read_csv('./data/adult.data', header=None)
df.head()
```

```python
with open('./data/adult.names', encoding='utf8') as f:
    data = f.readlines()
all_fields = []
skip_row = True
for row in data:
    if row.startswith('>50K'):
        skip_row = False
    if skip_row : continue
        
    fields=row.split(': ')
    if len(fields) == 2:
        all_fields.append(fields[0])
all_fields.append('Y')
df.columns = all_fields
df.head()
```

## 2. 資料清理、資料探索與分析

```python
df.info()
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        print(col)
        print(df[col].unique())
```

```python
df = df.replace(' ?', None)
for col in df.columns:
    if df[col].dtype == 'object':
        print(df[col].unique())
```

```python
df = df.dropna()
for col in df.columns:
    if df[col].dtype == 'object':
        dict1 = {}
        for i, category in enumerate(df[col].unique()):
            dict1[category]=i
        df.loc[:, col] = df[col].map(dict1)
    df[col] = df[col].astype(int)
df.head()
```

```python
df.info()
```

```python
# 是否有含遺失值(Missing value)
df.isnull().sum()
```

```python
df.to_csv('./clean_data.csv', index=False)
```

```python
y = df.Y
X = df.drop('Y', axis=1)
```

## 3. 不須進行特徵工程

## 4. 資料分割

```python
# 指定X，並轉為 Numpy 陣列
X = X.values

# 資料分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)

# 查看陣列維度
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 特徵縮放

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

## 5. 選擇演算法

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

## 6. 模型訓練

```python
clf.fit(X_train_std, y_train)
```

## 7. 模型評估

```python
y_pred = clf.predict(X_test_std)
y_pred
```

```python
# 計算準確率
print(f'{accuracy_score(y_test, y_pred)*100:.2f}%') 
```

```python
# 混淆矩陣
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# 混淆矩陣圖
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred))
disp.plot()
plt.show()
```
