# 計算及繪製混淆矩陣

## 載入資料

```python
y_true = [0, 0, 0, 1, 1, 1, 1, 1]
y_pred = [0, 1, 0, 1, 0, 1, 0, 1]
```

## 計算混淆矩陣

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_true, y_pred)
```

```python
# 依圖8.4順序顯示
from sklearn.metrics import confusion_matrix

confusion_matrix(y_true, y_pred, labels=[1, 0])
```

```python
# 取得混淆矩陣的4個格子
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
tn, fp, fn, tp
```

## 繪製混淆矩陣

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

ConfusionMatrixDisplay.from_predictions(
    y_true, y_pred, labels=[1, 0], display_labels=["true", "false"]
)
```

```python
# 方法 2
cm = confusion_matrix(y_true, y_pred, labels=[1, 0])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["true", "false"])
disp.plot()
```

```python
# 方法 3
fig, ax = plt.subplots(figsize=(5, 5))

# 顯示矩陣
ax.matshow(cm, cmap=plt.cm.Blues, alpha=0.3)

# 按 [1, 0] 順序
for i in range(cm.shape[0] - 1, -1, -1):
    for j in range(cm.shape[1] - 1, -1, -1):
        ax.text(x=j, y=i, s=cm[i, j], va="center", ha="center")

# 置換刻度
ax.set_xticks(range(cm.shape[0]), labels=["true", "false"], fontsize=14)
ax.set_yticks(range(cm.shape[1]), labels=["true", "false"], fontsize=14)

# 設定標籤
plt.xlabel("Predicted label", fontsize=16)
plt.ylabel("True label", fontsize=16)
```
