## 單變數分析(Univariate Analysis)

```python
sns.boxplot(x='GarageQual',y='SalePrice', data=df)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='KitchenQual',y='SalePrice',hue='CentralAir',data=df,split=True)
plt.show()
```

## 偵測離群值(Detect Outlier)

```python
from scipy import stats

q1, q3= np.percentile(df['SalePrice'], [25,75])
iqr = q3 - q1
lower_bound = q1 -(1.5 * iqr) 
upper_bound = q3 +(1.5 * iqr)
df['Zscore_SalePrice'] = stats.zscore(df['SalePrice'])
df[(df['Zscore_SalePrice'] < -3) | (df['Zscore_SalePrice'] > 3)]
```

## 移除離群值

```python
df2 = df.query('Zscore_SalePrice > -3 and Zscore_SalePrice < 3')
df2.head()
```

## 繪製散佈圖

```python
sns.scatterplot(x=df2['GarageArea'], y=df2['SalePrice'], hue=df2['Foundation'])
```

### price of houses depends upon the size of Garage Area they have but not that much on Foundation. 

## 繪製長條圖

```python
df.groupby(['GarageType','CentralAir'])['SalePrice'].mean().unstack(1).plot.barh()
```

## 處理離群值(Handling missing values)

```python
df2.loc[:, 'Exterior2nd'] = df2['Exterior2nd'].dropna()
```

```python
df2.loc[:, 'FireplaceQu'] = df2['FireplaceQu'].fillna(0)
```

```python
# 線性差補
# options: linear, time, index, values, nearest
df2.interpolate(method ='linear', limit_direction ='forward')
```

## 繪製熱力圖(HeatMap)

```python
sns.heatmap(df2.isnull())
plt.show()
```

```python
corr_matrix = df2[['MSSubClass','LotArea','MoSold']].corr()

# 顯示下三角
mask = np.zeros_like(corr_matrix)
# Returns copy of array with lower part of the triangle
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    sns.heatmap(corr_matrix, mask=mask, square=True, annot=True, fmt=".2f", center=0)
```
