## 複利計算
#### 存入10000元，固定利率5%，20年後會變成多少錢?
#### 提示：一年後，10000 * (1+0.05) = 10500

```python
import pandas as pd
from matplotlib import pyplot as plt

# Create a dataframe with 20 years
df = pd.DataFrame ({'Year': range(1, 21)})

# Calculate the balance for each year based on the exponential growth from interest
df['Balance'] = 10000 * (1.05**df['Year'])

#Display the dataframe
print(df)

# Plot the line

plt.plot(df.Year, df.Balance, color="green")
plt.xlabel('Year')
plt.ylabel('Balance')
plt.show()
```

## 固定利率15%，30年後會變成多少錢?

```python
print(f'{10000 * (1.15 ** 30):,.2f}')
```

## 固定利率1%，30年後會變成多少錢?
