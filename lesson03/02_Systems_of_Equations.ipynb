# 聯立方程式(Systems of Equations)
$\begin{equation}x + y = 16 \end{equation}$

$\begin{equation}10x + 50y = 560 \end{equation}$

## 方程式求解

```python
from sympy.solvers import solve
from sympy import symbols

x, y = symbols('x, y')
solve([x + y - 16, 10*x + 50*y - 560])
```

## 畫圖找出交叉點，即聯立方程式的解

```python
from matplotlib import pyplot as plt

# 極端值 1
dollar_10 = [16, 0]
dollar_50 = [0, 16]

# 極端值 2
dollar_10_2 = [56,0]
dollar_50_2 = [1,11]

# Plot the lines
plt.plot(dollar_10, dollar_50, color='blue')
plt.plot(dollar_10_2, dollar_50_2, color="orange")
plt.xlabel('x (10 dollar)')
plt.ylabel('y (50 dollar)')
plt.grid()

plt.show()
```

## 以 NumPy 解聯立方程式

```python
import numpy as np

a = np.array([[1, 1], [10, 50]])
b = np.array([16, 560])
print(np.linalg.solve(a, b))
```
